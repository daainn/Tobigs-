```python
import torch

import pandas as pd
import numpy as np

from sklearn.covariance import EllipticEnvelope
from sklearn.metrics import f1_score

import warnings
warnings.filterwarnings(action='ignore')
```

* 신용카드 사기거래 탐지 데이콘 대회에서 사용했던 알고리즘은 사이킷런에서 제공했던 `EllipticEnvelope`모델이다.

**EllipticEnvelope**

-
* 위사진에서 볼 수 있듯이 `EllipticEnvelope`은 데이터 세트 주위에 가상의 타원의 영역을 생성한 후 그 안에 들어오는 것을 일반데이터로 간주하고, 영역 밖에 위치하고 있는 점을 이상 데이터로 간주한다. 
* `EllipticEnvelope`는 가우시안 분포를 활용하여 사용하는 모델이다.
* 이번 대회에서는 `EllipticEnvelope` 모델만을 사용해서 이상탐지를 진행하였으며, 대회가 끝난 후 다른 팀의 모델들을 확인해본 결과 `EllipticEnvelope`과 다른 모델을 앙상블 하였을 때 더 좋은 성적을 거둔 것을 알 수 있었다.

그 이외에 사용한 `f1_score`은 `EllipticEnvelope`의 성능을 확인하기 위한 지표로 사용되었다.

```python
#train 데이터 불러오기
train_df = pd.read_csv('./train.csv') 
train_df.head()
```

```python
# Validation 데이터 불러오기
val_df = pd.read_csv('./val.csv')
val_df.head()
```

```python
# Validation 데이터셋의 사기 거래 비율이 다른 데이터셋에서도 비슷하게 발생할 것이라고 가정한 후 진행
val_normal, val_fraud = val_df['Class'].value_counts()
val_contamination = val_fraud / val_normal
print(f'Validation contamination : [{val_contamination}]')
```

```python
# Train dataset은 Label이 존재하지 않음
train_x = train_df.drop(columns=['ID']) # Input Data
```

```python
# 위에서 언급한 대로 Train dataset도 Validation dataset과 동일한 비율로 사기거래가 발생 했을 것이라고 생각하여 모델의 contamination을 validation contamination과 같게 설정
model = EllipticEnvelope(support_fraction = 0.994, contamination = val_contamination, random_state = 42)
model.fit(train_x)
```

```python
def get_pred_label(model, x, k):
  prob = model.score_samples(x)
  prob = torch.tensor(prob, dtype = torch.float)
  topk_indices = torch.topk(prob, k = k, largest = False).indices

  pred = torch.zeros(len(x), dtype = torch.long)
  pred[topk_indices] = 1
  return pred.tolist(), prob.tolist()

val_x = val_df.drop(columns=['ID', 'Class']) # Input Data
val_y = val_df['Class'] # Label

val_pred, val_prob = get_pred_label(model, val_x, 29)
val_score = f1_score(val_y, val_pred, average='macro')
print(f'Validation F1 Score : [{val_score}]')
```

* validation 데이터 셋으로 평가를 진행한 결과 f1-score의 macro average 점수가 0.92로 높게 측정된 것을 알 수 있다.

```python
#test 데이터셋 불러오기
test_df = pd.read_csv('./test.csv') 
test_df.head()
```

```python
test_x = test_df.drop(columns=['ID'])
```

```python
test_pred, _ = get_pred_label(model, test_x, 315)
print('n_fraud : ', sum(test_pred))
```
